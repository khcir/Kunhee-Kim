```python
# import BeautifulSoup, a Python library for parsing data out of HTML and XML files
from bs4 import BeautifulSoup

# send get request to URL
data = requests.get("https://www.influenster.com/reviews/lululemon")

# create an instance of BeautifulSoup
soup = BeautifulSoup(data.content, "html.parser")

# find all the "div" tabs with a specific class for the user written reviews
reviews = soup.find_all('div', class_ = "review-text sc-htoDjs knjVII sc-gzVnrw epkNeZ")

# notice the list is empty because the page has dynamic content
print(reviews)
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver


# create an instance of webdriver object, which is like opening a browser when the driver is called
# driver = webdriver.Chrome()
driver = webdriver.Chrome(executable_path='/Users/diyasmacbook/Desktop/chromedriver')

# navigate to the page
driver.get("https://www.influenster.com/reviews/lululemon")

# exceute Javascript inside the browser to scroll the webpage
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

# wait 15 seconds
import time
time.sleep(15)
```

```python
# attribute of where the source code of page is stored in a string format
data = driver.page_source 

# create an instance of BeautifulSoup
soup = BeautifulSoup(data, "html.parser")

# loop through to find all the div tags
for tag in soup.find_all("div"):
    
    # for prototype, see what type of tags we have
    
    # some tags won't have a CSS class, if so we check if it is the CSS class we want
    try:
        # print(tag["class"])
        # optional: if CSS class alias, bv, exists this specific webpage, print CSS class name
        # if 'bv' in tag["class"][0]:
        print(tag["class"])
        
    # otherwise continue to the next tag
    except:
        continue
```

```python
# for production in a .py file, use the code below

# have a try: 
# get review texts

# if div tag on this specific webpage refers to the CSS class, "bv-content-summary-body-text", print
for tag in soup.find_all("div", class_="review-text sc-htoDjs knjVII sc-gzVnrw epkNeZ"):
    print(tag.text)
    print()
```

```python
import pandas as pd

posts = [] # initialize dataframe

for tag in soup.find_all("div", class_="review-text sc-htoDjs knjVII sc-gzVnrw epkNeZ"):
    posts.append([tag.text])

posts = pd.DataFrame(posts, columns=['Review'])

print(posts)
```
